```
public class Product {
    private string name {get; set;}
    private decimal price {get; set;}
    private int quantity {get; set;}
    protected string category {get; set;}
    internal string description {get; set;}

    public Product(string name, decimal price, int quantity, string category, string description){
        this.name = name;
        this.price = price;
        this.quantity = quantity;
        this.category = category;
        this.description = description;
        
    }

    public decimal GetTotalPrice(){
        return quantity * price;
    }

    public void UpdateQuantity(int amount){
        quantity = amount;
    }

    public string GetProductInfo(){
        return $"Название товара: {name} \nСтоимость: {price} рублей \nКоличесвто: {quantity} штук \nКатегория: {category} \nОписание: {description}";
    }
}

public class Seller{
    private string name {get; set;}
    private string employeeId {get; set;}
    protected decimal salary {get; set;}
    internal string contactInfo {get; set;}

    public Seller(string name, string employeeId, decimal salary, string contactInfo){
        this.name = name;
        this.employeeId = employeeId;
        this.salary = salary;
        this.contactInfo = contactInfo;
    }

    private List<Product> products = new List<Product>();

    public void GetProductList(){
        foreach(var product in products){
            Console.WriteLine(product.GetProductInfo());
        }
    }

    public void AddProduct(Product product){
        products.Add(product);
    }

    public void SellProduct(Product product, int quantity){
        product.UpdateQuantity(quantity);
    }

    public string GetSellerInfo(){
        return $"имя продавца: {name} \nID продавца: {employeeId} \nЗарплата: {salary} рублей \nКонтактная информация: {contactInfo}";
    }
}

public class Store{
    private string storeName {get; set;}
    private string location {get; set;}
    public string storeHours {get; set;}

    public Store(string storeName, string location, string storeHours){
        this.storeName = storeName;
        this.location = location;
        this.storeHours = storeHours;
    }

    private List<Seller> sellers = new List<Seller>();

    public void AddSeller(Seller seller){
        sellers.Add(seller);
    }

    public void ListProducts(){
        foreach (var seller in sellers){
            seller.GetProductList();
        }
    }

    public string GetStoreInfo(){
        return $"Название магазина: {storeName} \nМестоположение: {location} \nЧасы работы: {storeHours}";
    }
}


Product product1 = new Product("Молоко", 80, 150, "Молочная продукция", "Вкусное молоко, 2.5% жирности");
Console.WriteLine("Стоимость всего количества товара: {0}", product1.GetTotalPrice());
product1.UpdateQuantity(20);
Console.WriteLine();
Console.WriteLine("информация о продукте:");
Console.WriteLine(product1.GetProductInfo());
Console.WriteLine();
Console.WriteLine();


Seller seller1 = new Seller("Иванов Иван Иванович", "1", 25000, "+7996123456");
seller1.AddProduct(product1);
Console.WriteLine("Информация о Продавце:");
Console.WriteLine(seller1.GetSellerInfo());
Console.WriteLine();


Store store1 = new Store("Продукты", "г.Тюмень, ул. Мельникайте, д. 70", "c 10 до 22");
store1.AddSeller(seller1);
Console.WriteLine("Товары магазина:");
store1.ListProducts();
Console.WriteLine();
Console.WriteLine("Информация о магазине:");
Console.WriteLine(store1.GetStoreInfo());
```
